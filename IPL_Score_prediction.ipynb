```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/IPL_scores.csv')
data.head()
```

```python
data.info()
```

```python
data.drop(['mid','striker','non-striker'], axis = 1, inplace = True)
```

```python
print(np.unique(data['bat_team']))
```

```python
all_teams=np.unique(data['bat_team'])
old_teams = ['Deccan Chargers', 'Kochi Tuskers Kerala', 'Gujarat Lions', 'Kochi Tuskers Kerala', 'Rising Pune Supergiant', 'Pune Warriors', 'Rising Pune Supergiants']
current_teams = [teams for teams in all_teams if teams not in old_teams]
data = data[data['bat_team'].isin(current_teams) & data['bowl_team'].isin(current_teams)]
```

```python
data.head()
```

```python
cat_cols = ['venue', 'bat_team', 'bowl_team', 'batsman','bowler']
for cols in cat_cols:
    n_unique = np.unique(data[cols]).shape[0]
    print("Number of unique values in %s column is %d"%(cols, n_unique))
```

```python
from sklearn.preprocessing import LabelEncoder
le_columns = ['batsman', 'bowler', 'venue']
le = LabelEncoder()
for col in le_columns:
  if col=='batsman':
    batsman_dict = dict(zip(data[col], le.fit_transform(data[col])))
    data[col] = le.fit_transform(data[col])
  elif col=='bowler':
    bowler_dict = dict(zip(data[col], le.fit_transform(data[col])))
    data[col] = le.fit_transform(data[col])
  elif col=='venue':
    venue_dict = dict(zip(data[col], le.fit_transform(data[col])))
    data[col] = le.fit_transform(data[col])
batsman_dict
```

```python
data
```

```python
nominal_cols = ['bat_team', 'bowl_team']
data = pd.concat((data, pd.get_dummies(data[nominal_cols])), axis = 1)
```

```python
data
```

```python
data.drop(nominal_cols, axis = 1, inplace = True)
data
```

```python
data.drop(['date'], axis = 1, inplace = True)
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(data.drop('total',axis=1),data['total'], train_size = 0.7)
```

```python
from sklearn.ensemble import RandomForestRegressor
random_forest_model = RandomForestRegressor()
random_forest_model.fit(x_train, y_train)
train_score_forest = str(random_forest_model.score(x_train, y_train)*100)
test_score_forest = str(random_forest_model.score(x_test, y_test)*100)
print(f'Train Score : {train_score_forest[:5]}%\nTest Score : {test_score_forest[:5]}%')
```
